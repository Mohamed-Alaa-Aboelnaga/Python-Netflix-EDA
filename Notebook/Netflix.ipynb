## Objective

### The goal of this project is to perform exploratory data analysis on Netflix data to understand movie duration patterns and analyze short action movies released in the 1990s.

## Import Libraries

### In this step, we import the libraries required for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

### Here we load the Netflix dataset into a pandas DataFrame.

```python
df = pd.read_csv("netflix_data.csv")
```

## Data Exploration

### We explore the dataset to understand its structure and check available columns.

```python
df.info()
```

```python
df.head()
```

```python
df.shape
```

## Filter Movies

### The dataset contains both movies and TV shows, here we filter only movies.

```python
movies = df[df["type"] == "Movie"]

movies
```

```python
movies_90s = df[(df["release_year"] >= 1990) & 
                    (df["release_year"] < 2000)]
movies_90s
```

```python
movies_90s["genre"].value_counts()
```

```python
movies_90s["release_year"].value_counts().sort_index()
```

## Distribution of Movie Durations in the 1990s

### This histogram shows the distribution of movie durations for movies released in the 1990s.

### The goal is to understand:
- The most common movie duration
- Whether most movies are short, medium, or long.

```python
sns.histplot(movies_90s["duration"])
plt.title("Distribution of Movie Durations in the 1990s")
```

## Number of Movies Released Each Year in the 1990s

### This visualization shows how many movies were released each year during the 1990s.

### It helps identify:
- Trends in movie production
- Years with higher or lower releases

```python
plt.figure(figsize=(8,5))

movies_90s["release_year"].value_counts().sort_index().plot(kind="line", marker="o")

plt.title("Number of Movies Released Each Year in the 1990s")
plt.xlabel("Year")
plt.ylabel("Number of Movies")

plt.show()
```

## Insights

### From the visualizations we observe:

- Most movies in the 1990s have durations around 90–120 minutes.
- The number of movie releases increased in the late 1990s.

## the most frequent movie duration in the 1990s

```python
duration = movies_90s["duration"].mode()[0] # [0] => to take the first value

print(duration)
```

```python
short_action_movies = movies_90s[(movies_90s["genre"] == "Action") & (movies_90s["duration"] < 90)]

short_action_movies
```

### The count of the number of short action movies released in the 1990s

```python
short_movie_count = len(short_action_movies)

short_movie_count
```
